```python
df['new_feature'] = df['rm'] * df['lstat']
df['new_feature']
```

```python
# Define age groups using boolean indexing
new_properties = df[df['age'] <= 35]
old_properties = df[(df['age'] > 35) & (df['age'] <= 70)]
very_old_properties = df[df['age'] > 70]

# Plot histograms for property age vs. median value
plt.figure(figsize=(12, 6))

plt.hist(new_properties['age'], bins=20, alpha=0.5, label='New Properties')
plt.hist(old_properties['age'], bins=20, alpha=0.5, label='Old Properties')
plt.hist(very_old_properties['age'], bins=20, alpha=0.5, label='Very Old Properties')

# plt.xlabel('Property Age')
# plt.ylabel('Frequency')
# plt.title('Distribution of Property Age') - x, y, title give errors like the past exercise.
plt.legend()

plt.show()
```

```python
bins = [0, 1, 5, 10, 20, max(df['crim'])]
labels = ['0-1', '1-5', '5-10', '10-20', '20+']
df['crim_bin'] = pd.cut(df['crim'], bins=bins, labels=labels)

# Group by binned crime rate and calculate average MEDV
avg_MEDV_per_CRIM_BIN = df.groupby('crim_bin')['medv'].mean()

# Plot line plot for average MEDV per binned crime rate
plt.figure(figsize=(10, 6))
plt.plot(avg_MEDV_per_CRIM_BIN.index, avg_MEDV_per_CRIM_BIN.values, marker='o')
# plt.xlabel('Binned Crime Rate')
# plt.ylabel('Average MEDV')
# plt.title('Average Property Value vs. Binned Crime Rate') Same Error: I tried changing the position of the str plt methods but they didn't work<\SORRY>
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
