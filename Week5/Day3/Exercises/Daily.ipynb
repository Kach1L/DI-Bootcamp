```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Create an array x of 100 numbers ranging from 0 to 20
x = np.linspace(0, 20, 100)

# Create an array y which is the square of x
y = x**2

# Plot x vs y
plt.plot(x, y, label='y = x^2')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Adjust the x and y limits
plt.xlim(0, 20)
plt.ylim(0, 400)

# Save the plot as a .png file
plt.savefig('xy_plot.png')
```

```python
# Load your dataset into a DataFrame (assuming you have a DataFrame named 'data')
data = pd.read_csv('chipo.txt', sep = '\t')

# 2. Clean the item_price column and transform it into a float
data['item_price'] = data['item_price'].apply(lambda x: float(x.replace('$', '')))

# 3. Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['item_price'], data['quantity'])

# 4. Add title and labels
plt.title('Relationship between Price of Items and Quantity Ordered')
plt.xlabel('Item Price ($)')
plt.ylabel('Quantity Ordered')

# 5. Save the plot as a .png file
plt.savefig('scatter_plot.png')
```

```python
# Load the 'chipo' data into a DataFrame (assuming you have a DataFrame named 'chipo')
chipo = pd.read_csv('chipo.txt', sep = '\t')

# 1. Group the data by item_name and calculate the total quantity for each item
item_quantities = chipo.groupby('item_name')['quantity'].sum().reset_index()

# 2. Create a bar plot for the top 5 items
top_items = item_quantities.nlargest(5, 'quantity')
plt.figure(figsize=(10, 6))
plt.bar(top_items['item_name'], top_items['quantity'])

# 3. Add title and labels
plt.title('Top 5 Items by Total Quantity Ordered')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity Ordered')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# 4. Save the plot as a .png file
plt.savefig('top_items_bar_plot.png')
```
