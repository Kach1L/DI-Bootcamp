```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

figure, axes = plt.subplots(1,2)
axes[0].set(title='y = sin(x)')
axes[1].set(title='y = cos(x)')
line_plot1 = axes[0].plot(x,y1, color='tab:red', marker='')
line_plot2 = axes[1].plot(x,y2, color='tab:blue', marker='o')

plt.tight_layout()
```
